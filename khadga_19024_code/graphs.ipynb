```python
b = [1,3,6]
x = np.arange(0,1.1,0.1)
def y_(b):
    return 1 - m.e**(-b*(1-x))
# plt.xticks(x)
plt.xlim((0,1))
plt.ylim((0,1))
for i in b:
    plt.plot(x,y_(i))
plt.title(r"$Weight_{CE}\ =\ \left(1-e^{-\beta\left(1-p^s_t\right)}\right)$")
plt.legend([r'$\beta$ =' + str(i) for i in b])
plt.xlabel(r"$p^t_s$")
plt.ylabel(r"$weight_{CE}$")
```
